```python
# Testing some data.
# There are two many points. The magnetic field variation was too slow
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the CSV file into a DataFrame
# Replace 'your_data.csv' with your actual filename
df = pd.read_csv("data/y_by_time.csv")
df2 = df.assign(time=df['second']+df['nsec']/10E9,value=df['value'])

# 2. Optional: Preview your data structure
print(df.head())
print(df2.head())

# 3. Create the plot
plt.figure(figsize=(10, 5))  # Sets the display size

# Replace 'X_Column' and 'Y_Column' with your exact CSV column headers
plt.plot(df2["time"], df2["value"],linestyle="-", color="r")

# 4. Add details and labels
plt.title("Y sensor position vs time")
plt.xlabel("time")
plt.ylabel("Y Axis position")
plt.grid(True)

# 5. Render the graph directly inside your notebook
plt.show()
```
